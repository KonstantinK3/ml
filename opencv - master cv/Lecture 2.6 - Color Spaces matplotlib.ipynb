## Let's take a closer look at color spaces

You may have remembered we talked about images being stored in RGB (Red Green Blue) color Spaces. Let's take a look at that in OpenCV.

### First thing to remember about OpenCV's RGB is that it's BGR (I know, this is annoying)

Let's look at the image shape again. The '3L' 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread('./images/input.jpg')
```

### Let's look at the individual color levels for the first pixel (0,0)

```python
# BGR Values for the first 0,0 pixel
B, G, R = image[10, 50] 
print (B, G, R)
print (image.shape)
```

Let's see what happens when we convert it to grayscale

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print (gray_img.shape)
print (gray_img[10, 50])
```

It's now only 2 dimensions. Each pixel coordinate has only one value (previously 3) with a range of 0 to 255

```python
gray_img[0, 100] 
```

### Another useful color space is HSV 
Infact HSV is very useful in color filtering.

```python
#Hue/Farbton: 0 - 180, Saturation: 0 - 255, Value: 0 - 255

image = cv2.imread('./images/input.jpg')

hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

cv2.imshow('HSV image', hsv_image)
# Die Doppelpunkte wählen das Gesamte Bild aus x-y Dimensionen des Graubildes. Der letzte Parameter wählt den Channel
cv2.imshow('Hue channel', hsv_image[:, :, 0])
cv2.imshow('Saturation channel', hsv_image[:, :, 1])
cv2.imshow('Value channel', hsv_image[:, :, 2])# Value Channel ist im Prinzip ein Graubild

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print('hsv image')
plt.imshow(hsv_image[:,:,[2,1,0]])
```

```python
print('Hue channel')
plt.imshow(hsv_image[:,:,[2,1,0]][:, :, 0])
```

```python
print('Saturatuion channel')
plt.imshow(hsv_image[:,:,[2,1,0]][:, :, 1])
```

```python
print('Value channel')
plt.imshow(hsv_image[:,:,[2,1,0]][:, :, 2])
```

### Let's now explore lookng at individual channels in an RGB image

```python
image = cv2.imread('./images/input.jpg')

# OpenCV's 'split' function splites the image into each color index
B, G, R = cv2.split(image)

print (B.shape)
plt.imshow(R)
```

```python
plt.imshow(G)
```

```python
plt.imshow(B)
```

```python
R.shape
```

```python
# Let's re-make the original image, 
merged = cv2.merge([B, G, R])
plt.imshow(merged[:,:,[2,1,0]]) 
```

```python
# Let's amplify the red color
merged = cv2.merge([B, G, R+100])
plt.imshow(merged[:,:,[2,1,0]]) 
```

```python
import cv2
import numpy as np

B, G, R = cv2.split(image)

# Let's create a matrix of zeros 
# with dimensions of the image h x w  
zeros = np.zeros(image.shape[:2], dtype = "uint8")

print('Red')
plt.imshow(cv2.merge([zeros, zeros, R])[:,:,[2,1,0]])


'''cv2.imshow("Green", cv2.merge([zeros, G, zeros]))
cv2.imshow("Blue", cv2.merge([B, zeros, zeros]))

cv2.waitKey(0)
cv2.destroyAllWindows()'''
```
