```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve
%matplotlib inline
```

```python
def f(x):
    return np.sin(x/5.)*np.exp(x/10.)+5*(np.exp(-x/2.))
```

```python
x = np.arange(0, 16, 0.1)
y = f(x)
plt.plot(x,y)
plt.show()
```

просто пример многочлена на графике

```python
def mnogo1(x):
    return 10 + 5 * x - 0.2 * x**2 + 0.01 * x**3
```

```python
x = np.arange(0, 16, 0.1)
y = f(x)
plt.plot(x,y)
y2 = mnogo1(x)
plt.plot(x,y2)
plt.show()
```

```python
polynom_rank = 1
A_1 = [
    [1 ** n for n in range(0, polynom_rank + 1)],
    [15 ** n for n in range(0, polynom_rank + 1)]
]

b_1 = [f(1), f(15)]
solve(A_1, b_1)
```

```python
def mnogo1(x):
    return 3.43914511 - 0.18692825*x
```

```python
x = np.arange(0, 16, 0.1)
y = f(x)
plt.plot(x,y)
y2 = mnogo1(x)
plt.plot(x,y2)
plt.show()
```

```python
polynom_rank = 2
A_2 = [
    [1 ** n for n in range(0, polynom_rank + 1)],
    [8 ** n for n in range(0, polynom_rank + 1)],
    [15 ** n for n in range(0, polynom_rank + 1)]
]

b_2 = [f(1), f(8), f(15)]
w1, w2, w3 = solve(A_2, b_2)
```

```python
def mnogo2(x, w1, w2, w3):
    return w1 + w2*x + w3*x**2
```

```python
x = np.arange(0, 16, 0.1)
y = f(x)
plt.plot(x,y)
y2 = mnogo2(x, w1, w2, w3)
plt.plot(x,y2)
plt.show()
```

```python
polynom_rank = 3
A_3 = [
    [1 ** n for n in range(0, polynom_rank + 1)],
    [4 ** n for n in range(0, polynom_rank + 1)],
    [10 ** n for n in range(0, polynom_rank + 1)],
    [15 ** n for n in range(0, polynom_rank + 1)]
]

b_3 = [f(1), f(4), f(10), f(15)]
w1, w2, w3, w4 = solve(A_3, b_3)
```
