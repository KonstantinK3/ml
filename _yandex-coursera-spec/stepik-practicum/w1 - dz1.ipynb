```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from math import atan
```

```python
def f(x):
    return 2*atan(x)
```

```python
xs = np.linspace(-10,10,100)

ys = []

for y in xs:
    ys.append(f(y))

plt.plot(xs, ys)
plt.show()
```

```python
lim = round(f(float('inf')), 3)
print(lim)
```

```python
from math import exp
```

```python
def f(x):
    return exp(x)
```

```python
xs = np.linspace(-10,10,100)

ys = []

for y in xs:
    ys.append(f(y))

plt.plot(xs, ys)
plt.show()
```

```python
def f(x):
    return (2*x**2-3*x-5)/(3*x**2+x+1)
```

```python
xs = np.linspace(-10,10,100)

ys = []

for y in xs:
    ys.append(f(y))

plt.plot(xs, ys)
plt.show()
```

```python
round(f(1e20),3)
```

```python
from math import pi, sin
def f(x):
    return (sin(pi*x/2)/x)
```

```python
xs = np.linspace(-30,30,100)

ys = []

for y in xs:
    ys.append(f(y))

plt.plot(xs, ys)
plt.show()
```
